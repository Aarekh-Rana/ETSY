```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Load data
train_df = pd.read_parquet(r'C:\Users\aarek\OneDrive\문서\DCU\ETSY\data\data\2024\train')
test_df = pd.read_parquet(r'C:\Users\aarek\OneDrive\문서\DCU\ETSY\data\data\2024\test')

# Feature selection and preprocessing
# As a simple start, we'll use textual data from 'title' and 'tags' combined as our features
# More sophisticated text processing and other features can be added later
train_df['text'] = train_df['title'] + ' ' + train_df['tags']
test_df['text'] = test_df['title'] + ' ' + test_df['tags']

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=1000)  # limiting to the top 1000 features

X_train = vectorizer.fit_transform(train_df['text'])
y_train = train_df['top_category_id']
X_test = vectorizer.transform(test_df['text'])

# Model training
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
predictions = model.predict(X_test)

# Evaluation (Assuming you have labels for the test set, otherwise this step is not possible directly)
# y_test = test_df['top_category_id']  # Uncomment if test labels are available
# print(classification_report(y_test, predictions))

# Save model output for submission or further evaluation
output_df = pd.DataFrame({'product_id': test_df['product_id'], 'predicted_top_category_id': predictions})
output_df.to_csv(r'C:\Users\aarek\OneDrive\문서\DCU\ETSY\data\predictions_top_category.csv', index=False)
```
